```python
X = pd.get_dummies(train_df['Sex']).values
print(X)
```

```python
y = pd.get_dummies(train_df['Survived']).values
print(y)
```

```python
from sklearn.model_selection import train_test_split
X_train ,X_test, Y_train,Y_test = train_test_split(X,y, test_size=0.2)
print(X_train)
```

```python
svc = SVC()
```

```python
svc.fit(X_train.reshape(-1,1), Y_train.ravel())
```

```python
Y_pred = svc.predict(X_test.reshape(-1,1))
```

```python
acc_svc = round(svc.score(X_train.reshape(-1,1), Y_train.ravel()) * 100, 2)
print(acc_svc)
```

```python
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(X_train, Y_train)
Y_pred = knn.predict(X_test)
acc_knn = round(knn.score(X_train, Y_train) * 100, 2)
acc_knn
```
